```python
import pandas as pd 
import numpy as np
```

```python
covid_db = pd.read_csv("C:/Users/rnikk/OneDrive/Desktop/git_projects/covid_19_india.csv" )
```

```python
covid_db
```

```python
covid_db.drop("Sno",inplace = True , axis = 1)
```

As we can see there is no useful data available in ConfirmedIndianNational,ConfirmedForeignNational and time columns we must drop them in order to make our data look more cleaner

```python
covid_db.drop(["ConfirmedIndianNational","ConfirmedForeignNational","Time"],inplace = True , axis =1)
```

```python
covid_db.head()
```

Lets see the total number of Cured, Deaths and Confirmed cases in each state(using group by)

```python
covid_db.groupby("State/UnionTerritory").sum().sort_values("Confirmed",ascending = False).head()
```

Now, lets the the total number of active cases in each state(using pivot_table fuction)

```python
covid_db['Active_cases']=covid_db['Confirmed']-(covid_db['Cured']+covid_db['Deaths'])
```

```python
covid_db.head()
```

```python
statewise = pd.pivot_table(covid_db, values=["Confirmed","Cured","Deaths","Active_cases"],
                            index = "State/UnionTerritory",aggfunc = max)
```

```python
statewise.head()
```

lets see the recovery and mortality rates of each state

```python
statewise['Recovery_rate']=statewise['Cured']*100/statewise['Confirmed']
```

```python
statewise['Mortality_rate']=statewise['Deaths']*100/statewise['Confirmed']
```

```python
statewise=statewise.sort_values("Confirmed",ascending = False)
```
